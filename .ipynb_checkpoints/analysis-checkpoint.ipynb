```python
import pandas as pd 
import pickle
import os
```

```python
dict_kernel = {
    'linear' : 'Linear Kernel',
    'rbf' : 'RBF Kernel',
    'matern' : 'MaternKernel',
    'poli1' : 'Polynomial Kernel with Power = 1',
    'poli2' : 'Polynomial Kernel with Power = 2',
    'rq' : 'RQ Kernel',
    'preio' : 'Periodic Kernel',

}
```

```python
df_list = []
for path in os.listdir("src/pickles/"):
    if (path.split("_")[0] == "results"):
        optimazer = path.split("_")[1]
        kernel = dict_kernel[path.split("_")[2].split(".")[0]]
        results = pickle.load(open(f"src/pickles/{path}", "rb"))
        for result in results:
            method = result[0].split()[0]
            dataset = result[0].split()[2]
            mse = result[0].split()[10]
            standard_deviation = result[0].split()[12]
            rst = result[0].split()[10:13]
            train_time = result[2] - result[1]
            all_time = result[3] - result[1]

            df_list.append([optimazer,kernel,method,dataset,mse,standard_deviation,rst,train_time,all_time])
```

```python
df_coll = ["Optimazer", "Kernel", "Method", "DataSet", "MSE", "Standard Deviation", "Result", "Train Time", "All Process"]

df = pd.DataFrame(df_list, columns=df_coll)
```
